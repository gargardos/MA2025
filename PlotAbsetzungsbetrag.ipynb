```python
import matplotlib.pyplot as plt
import numpy as np

# Definitionsbereiche und zugehörige Funktionswerte
x1 = np.linspace(0.01, 99.99, 100)
y1 = x1

x2 = np.linspace(100.01, 519.99, 100)
y2 = 0.2 * x2 + 80

x3 = np.linspace(520.01, 999.99, 100)
y3 = 0.3 * x3 + 28

x4 = np.linspace(1000.01, 1199.99, 100)
y4 = 0.1 * x4 + 228

# Erstellen des Plots
plt.figure(figsize=(10, 6))

# Zeichnen der Funktionen mit kontrastreichen Farben
plt.plot(x1, y1, label='f(x) = x', color='blue', linewidth=2)
plt.plot(x2, y2, label='f(x) = 0,2x + 80', color='red', linewidth=2)
plt.plot(x3, y3, label='f(x) = 0,3x + 28', color='green', linewidth=2)
plt.plot(x4, y4, label='f(x) = 0,1x + 228', color='purple', linewidth=2)

# Achsenbeschriftungen
plt.xlabel("anzurechnendes Einkommen in €")
plt.ylabel("Absetzungsbetrag in €")

# Achsen-Skalierung
plt.xticks(np.arange(0, 1300, 100))
plt.ylim(0, 400)
plt.gca().set_aspect('equal', adjustable='box')  # gleichmäßige Skalierung

# Gitter & Legende
plt.grid(True)
plt.legend()

# Layout und Anzeige
plt.tight_layout()
plt.show()
```
